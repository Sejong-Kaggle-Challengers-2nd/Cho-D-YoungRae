# 데이터 전처리

## OC
1. open: 1, close: 0 라벨링

```python
train_df['OC'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# fit_transform 의 인자로 trian[['OC']]를 넣는다. train_df['OC'] X 
# DataFrame이 반환되냐, Series가 반환되냐 의 차이
type(train_df['OC'])
type(train_df[['OC']])
OC_encoder = OrdinalEncoder()
OC_encoded = OC_encoder.fit_transform(train_df[['OC']])
train_df['OC'] = OC_encoded
```

```python
train_df['OC'].value_counts()
```

```python
# 피처와 타겟값 분리
y = train_df['OC']
X = train_df.drop('OC', axis=1)
```

```python
X.info()
```

## ownerChange
1. 비어있는 값 -> 'unknown' 로 새로운 값 부여 -> 원핫인코딩

```python
from sklearn.preprocessing import OneHotEncoder

X['ownerChange'][X['ownerChange'].isnull()] = 'unknown'
ownerChange_encoder = OneHotEncoder()
ownerChange_encoded = ownerChange_encoder.fit_transform(X[['ownerChange']])
X['ownerChange'] = ownerChange_encoded
```

## instkind
1. 비어있는 값 -> 가장 많은 값으로 -> 원핫인코딩

```python
X['instkind'][X['instkind'].isnull()] = X['instkind'].value_counts().index[0]
instkind_encoder = OneHotEncoder()
instkind_encoded = instkind_encoder.fit_transform(train_df[['instkind']])
train_df['instkind'] = instkind_encoded
```

## sido, sgg
1. 원핫인코딩

```python
sido_encoder = OneHotEncoder()
sgg_encoder = OneHotEncoder()

sido_encoded = sido_encoder.fit_transform(train_df[['sido']])
sgg_encoded = sgg_encoder.fit_transform(train_df[['sgg']])


train_df['sido'] = sido_encoded
train_df['sgg'] = sgg_encoded
```

## 그외 숫자 피처들
1. 비어있는 값 -> 평균값

```python
num_attribs = ['openDate', 'bedCount',
       'revenue1', 'salescost1', 'sga1', 'salary1', 'noi1', 'noe1',
       'interest1', 'ctax1', 'profit1', 'liquidAsset1', 'quickAsset1',
       'receivableS1', 'inventoryAsset1', 'nonCAsset1', 'tanAsset1',
       'OnonCAsset1', 'receivableL1', 'debt1', 'liquidLiabilities1',
       'shortLoan1', 'NCLiabilities1', 'longLoan1', 'netAsset1', 'surplus1',
       'revenue2', 'salescost2', 'sga2', 'salary2', 'noi2', 'noe2',
       'interest2', 'ctax2', 'profit2', 'liquidAsset2', 'quickAsset2',
       'receivableS2', 'inventoryAsset2', 'nonCAsset2', 'tanAsset2',
       'OnonCAsset2', 'receivableL2', 'debt2', 'liquidLiabilities2',
       'shortLoan2', 'NCLiabilities2', 'longLoan2', 'netAsset2', 'surplus2',
       'employee1', 'employee2']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
  ('imputer', SimpleImputer(strategy='mean'))                        
])
```

```python
from sklearn.compose import ColumnTransformer

col_transformer = ColumnTransformer([
  ('num', num_pipeline, num_attribs)                                     
])

X_processed = col_transformer.fit_transform(X)
```

```python
X_processed.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rnd_clf = RandomForestClassifier(n_jobs=-1)
accuracy = cross_val_score(rnd_clf, X_processed, y, scoring='accuracy')
```
